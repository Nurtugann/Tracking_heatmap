```python
import requests
import json

TOKEN = "c611c2bab48335e36a4b59be460c57d2BF8416B73C4A65F2B8A88A5848E97CD4471F14C6"
BASE_URL = "https://hst-api.wialon.host/wialon/ajax.html"

def login(token):
    r = requests.get(BASE_URL, params={
        "svc": "token/login",
        "params": json.dumps({"token": token})
    })
    return r.json().get("eid")

def get_items(sid, item_type, flags):
    r = requests.get(BASE_URL, params={
        "svc": "core/search_items",
        "params": json.dumps({
            "spec": {
                "itemsType": item_type,
                "propName": "sys_name",
                "propValueMask": "*",
                "sortType": "sys_name"
            },
            "force": 1,
            "flags": flags,
            "from": 0,
            "to": 0
        }),
        "sid": sid
    })
    return r.json().get("items", [])

if __name__ == "__main__":
    SID = login(TOKEN)
    units = get_items(SID, "avl_unit", 1)
    if not units:
        print("Юниты не найдены.")
    else:
        print("Список юнитов:")
        for unit in units:
            print(f"{unit['id']}: {unit['nm']}")
```
